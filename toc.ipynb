```python
import os
from datetime import datetime
from IPython.display import Markdown, display

from pathlib import Path
import sys
```

```python
# Global variables
HERE = Path().resolve()
print(f"HERE: {HERE}")
ROOT = HERE.parent
print(f"ROOT: {ROOT}")
DATA = ROOT / 'data'
print(f"DATA: {DATA}")
```

```python
# Path to the notebooks directory
# notebooks_dir = os.path.dirname(os.path.abspath("__file__"))
notebooks_dir = HERE / 'notebooks'
print(f"notebooks_dir: {notebooks_dir}")
```

```python
# Function to generate the table of contents
def generate_toc(directory):
    # Check if repositorty exists
    if not os.path.exists(directory):
        return f"⚠️ The directory '{directory}' does not exist."

    # Lists all the .ipynb files in the directory
    notebooks = [
        f for f in os.listdir(directory)
        if f.endswith('.ipynb') and not f.startswith('.')
    ]

    if not notebooks:
        return "⚠️ No notebooks found in this directory."

    # Generate a Markdown entry for each notebook
    md_content = ["# Table of Contents of notebooks\n"]
    md_content.append(f"*Last updated: {datetime.now().strftime('%d/%m/%Y %H:%M')}*\n")

    for notebook in sorted(notebooks):
        chemin = os.path.join(directory, notebook)
        # Date of last modification
        date_modif = datetime.fromtimestamp(os.path.getmtime(chemin)).strftime('%d/%m/%Y')
        # Clickable link (works in Jupyter and Binder)
        lien = f"[{notebook}]({chemin})"
        md_content.append(f"- {lien} *(modified on {date_modif})*")

    return "\n".join(md_content)
```

```python
# Display the TOC
display(Markdown(generate_toc(notebooks_dir)))   
```
